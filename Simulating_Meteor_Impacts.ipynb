### Calculating Impact Velocity and Forces

We can use Euler's method in order to estimate the changing velocity in the x and y components as the meteorite crashes into earth with an initial velocity, height, and angle with respect to earth

$$ V_{new} = V_{old} + {dv \over dt} \times dt $$

To find $dv \over dt$ we use Newton's second low of motion:

$$ a = f/m $$
$$ Fnet_y = F_g - F_{drag_y} $$
$$ Fnet_x = - F_{drag_x} $$

The equation for drag force:

$$ Fdrag = 0.5 \times C_d \times \rho \times A \times v^2 $$
$$ Fdrag_y = 0.5 \times C_d \times \rho \times A \times {v_y}^2 $$
$$ Fdrag_x = 0.5 \times C_d \times \rho \times A \times {v_x}^2 $$

Where $ v_x = v_{initial} \times sin\theta $ and $ v_y = v_{initial} \times cos\theta $

Putting this all together:

**In the y direction:**

The changing velocity is given by:
$$ V_{new} = V_{old} + (g - F_{drag_y}) \times dt$$

The chaning height is given by updating the new height of the ball after it falls some distance $v_y \times d_t$
$$ h_{new} = h_{old} - V_{new} \times dt $$

**In the x direction:**
Instead of calculating it based on time, we calculate it based on the amount of iterations it took the meteor to reach height = 0 in the y direction Eulers method. That being said we still use a timestep of 0.1 to mimic updating the velocity with respect to time, the iterations just represents the cutoff for the iterations.

The changing velocity is given by:
$$ V_{new} = V_{old} - F_{drag_y} \times dt$$

The chaning position is given by updating the new distance of the ball after it travels some distance $v_x \times d_t$
$$ x_{new} = x_{old} + V_{new} \times dt $$

```python
import numpy as np
import matplotlib.pyplot as plt

#Initial conditions

dt = 0.1 # s
g = 9.81 #m/s
tmax = 100 # s
steps = int(tmax / dt) # Total number of steps
t = [0.] # List to store time values corresponding to position/velocity values
m = 100000 # kg, must be large to impact earth
r = 1.87 # m, based on the mass of the meteor we can assume this
A = np.pi * (r **2)
V = 4/3*(np.pi)*(r**3)
p = m/V
rho = 1.225  # air density kg/m^3
Cd = 0.47
x_initial = 0
delta_t = 0.1 #Assumed impact duration

# initial Velocities
vinitial = 20000 # m/s
theta = 30 #assuming the angle of entry is at a 45 degree angle here
cos_theta_rad = np.cos(np.radians(theta))
sin_theta_rad = np.sin(np.radians(theta))

# Initial Velocities based on the angle the meteor enters at
vi_y = 20000 * sin_theta_rad
vi_x = 20000 * cos_theta_rad


h_initial = 100000 # m, altitude entered at 
h_i = h_initial 
#When h_initial goes into the smaller while loop, it goes down to become 0, so if we want to increment 
#... larger and larger heights, we need something to keep track of the height dropped at
time = 0


distance_array = []
v_x_array = []

height_array = []
v_y_array = []

iterations = 0

while h_initial > 0 :#Since we reiterate every time to find the velocity at a given moment until the height reaches zero at which this is the highest velocity
    
        drag_force = (Cd * rho * A * vi_y**2) / (2 * m)
    
        v_new = vi_y + (g - drag_force)*dt #updating the force down
    
        h_new = h_initial - (v_new * dt) #Updating the new height of the ball after it falls some distance v * d_t 

        if h_new < 0 :
            h_new = 0

        height_array.append(h_new)
        v_y_array.append(v_new)
              
        vi_y = v_new
        h_initial = h_new
        time += dt 

        # Count iterations
        iterations += 1

print(f"Vertical impact velocity: {vi_y:10.2f} m/s")

for i in range(iterations):  # Use the same number of iterations
    # Calculate drag force for the x-direction
    drag_force_x = (Cd * rho * A * vi_x**2) / (2 * m)

    # Update horizontal velocity
    v_new_x = vi_x - (drag_force_x * dt)

    # Update horizontal position
    x_new = x_initial + (v_new_x * dt)

    distance_array.append(x_new)
    v_x_array.append(v_new_x)

    # Prepare for next iteration
    vi_x = v_new_x  # Update horizontal velocity for the next iteration
    x_initial = x_new  # Update horizontal position for the next iteration

# After the loop
print(f"Horizontal impact velocity: {v_new_x:8.2f} m/s")

v_impact = np.sqrt( (v_new_x)**2 + (vi_y)**2)
print(f"Total Impact of velocity of: {v_impact:7.2f} m/s")

print(f"Horizontal distance traveled: {x_new:.2f} m")

# Impact force calculation
impact_force = (m * v_impact) / delta_t
print(f"Force upon impact {impact_force:24.2f} N")

# Plot the changes in horizontal distance and horizontal velocity

plt.scatter(distance_array, v_x_array, color='blue')
plt.xlabel('Horizontal Distance (m)')
plt.ylabel('Horizontal Velocity (m/s)')
plt.title('Change in Displacement of Meteorite vs Horizontal Velocity Over Time')
plt.show()

# Plot the changes in height and vertical velocity
plt.scatter(v_y_array, height_array, color='blue')
plt.xlabel('Vertical Velocity (m/s)')
plt.ylabel('Height (m)')
plt.title('Change in Height of Meteorite vs. Vertical Velocity over time')
plt.show()
```

#### Side Note: Newton's Experimental Law:

Newton's experimental law relates to the **coefficient of restitution (e)**, a measure of how "elastic" a collision is. In simple terms, it tells us how much kinetic energy is conserved during a collision. The equation for the coefficient of restitution is:

$$ e = \frac{|\text{Relative velocity of separation after collision}|}{|\text{Relative velocity of approach before collision}|} = \sqrt{\frac{\text{Final kinetic energy}}{\text{Initial kinetic energy}}} $$

This coefficient determines how much energy is lost upon impact. A value of $e = 1$ corresponds to a perfectly elastic collision where no kinetic energy is lost, while $e = 0$ represents a completely inelastic collision where the objects do not separate after impact. All real-world imoaced object have a coefficient of resitution.

In the context of our meteor impact simulation, we are using a **coefficient of restitution (e)** of **0.1**. This simulates a highly inelastic collision, such as a meteor impacting rocky terrain, where significant energy is lost due to the deformity and destruction caused by the impact.

We can calculate the final velocity after the collision using the relationship derived from Newton’s law. Starting with the initial velocity of the meteor upon impact *v_impact* and using the coefficient of restitution, we can find the final velocity *v_final* as follows:

$$ v_{final} = e \times v_{impact} $$

For example, if the meteor impacts the ground with a velocity of 852.78 m/s and we use $e = 0.1$, the final velocity after the collision would be:


$$ v_{final} = 0.1 \times 852.78 = 85.28 \, \text{m/s} $$


This reduced velocity is crucial to calculating the speed at which the meteor rolls or moves post-impact, having lost most of its kinetic energy during the collision.


### Calculating Rolling Distance Upon Impact
- As the $ 30 \degree $ does not exceed the maximum impact velocity of 1000m/s, which we will derive in the next experiment, the meteor will roll a certain distance upon impact

**Initial Kinetic Energy Before Impact**:
   The initial kinetic energy of the meteor before impact is given by:
   
   $$E_{initial} = \frac{1}{2} m v_{impact}^2 $$
   where:
   - m is the mass of the meteor,
   - $ v_{impact} $ is the velocity at the moment of impact.

**Energy Lost Due to Penetration**:
   Upon impact, the meteor penetrates the surface to a depth $ d_{penetration} $, during which it experiences a resistive force $F_{penetration}$. The work done by this resistive force reduces the meteor's kinetic energy:
   
   $$ W_{penetration} = E_{lost\_penetration} = F_{penetration} \times d_{penetration} $$

- Where $E_{lost\_penetration} = f_{penetration} \times d_{penetration} $
   
   This represents the energy lost due to penetration into the ground.

**Remaining Kinetic Energy After Penetration**:
   After losing energy to penetration, the remaining kinetic energy of the meteor is:
   
   $$E_{remaining} = E_{final} = \frac{1}{2} m v_{final}^2 - E_{lost\_penetration} $$
   
   where:
   - $ v_{final} is the meteor’s velocity after penetration (in our case, related to the impact velocity via the coefficient of restitution).

**Frictional Force During Rolling**:
   As the meteor rolls, the frictional force $ F_{friction} $ opposes its motion, causing a gradual reduction in its kinetic energy. The frictional force is given by:
   
   $$ F_{friction} = \mu_r \times m \times g $$

   where:
   - $ mu_r $ is the coefficient of rolling friction,
   - $ g $ is the acceleration due to gravity.

**Updating Kinetic Energy with Euler’s Method**:
   The meteor continues to roll, losing energy due to friction. The remaining kinetic energy after a small distance increment $ \Delta d$ is:

   $$ E_{kinetic, new} = E_{remaining} - F_{friction} \times \Delta d $$
   $$ E_{kinetic, new} = E_{kinetic, old} - F_{friction} \times \Delta d $$
   
   where $ \Delta d $ is the distance traveled in a small time step $ dt $, and can be approximated using the current velocity:
   
   $$ \Delta d = v \times dt $$
   
   Here, $ v $ is the velocity based on the remaining kinetic energy:
   
   $$ v = \sqrt{\frac{2 E_{\text{kinetic}}}{m}} $$
   

**Iterative Calculation of Rolling Distance**:
   Using Euler’s method, we update the kinetic energy and calculate the new distance traveled during each small time step. The total rolling distance is obtained by summing the small increments $ \Delta d $ at each step until the kinetic energy is fully dissipated ($ E_{kinetic} = 0 $)

Thus, the rolling distance is computed iteratively by repeatedly applying:

   $$E_{kinetic, new} = E_{kinetic, old} - E_{friction}  $$
   $$E_{kinetic, new} = E_{kinetic, old} - F_{friction} \times v \times dt $$
   
   and updating the distance $ d $ as:
   
   $$ d_{new} = d_{old} + \Delta d $$

This process continues until the remaining kinetic energy is fully dissipated due to frictional forces.

#### Calculating Impact Force Calculation

- **Mass (m)** = 100,000 kg  
- **Radius (r)** = 1.87 m  
- **Cross-sectional Area (A)** = 
- A = $\pi r^2 = \pi (1.87)^2 = 10.99 \, \text{m}^2$
  

From our assumption, we assume the meteorite is made of a strong stone material typical in many meteors, with an **Ultimate Tensile Strength (UTS)** of **90 MPa**This represents a dense stone material or iron/nickel alloy.

The formula for force in terms of the material strength is given by:


$$ F_{destroy} = \text{UTS} \times A $$


Substituting the values:


$$ F_{\text{destroy}} = 90 \times 10^6 \times 10.99 = 989.1 \times 10^6 N  $$

Thus, the meteor will be destroyed if the force exceeds  **989.1 Million Newtons**.



#### Critical Velocity Calculation

We can now use momentum to calculate the **critical velocity** at which the meteor will be destroyed. The force during impact is given by:


$$ F_{impact} = \frac{m \times v_{\text{critical}}}{\Delta t} $$

where:
- $ m = 100,000 \text{kg} $ is the mass of the meteor,
- $ v_{critical}$ is the critical velocity (which we want to find),
- $ \Delta t = 0.1s $ is the assumed impact duration.

Rearranging the formula to solve for $ v_{critical}:$


$$ v_{critical} = \frac{F_{\text{destroy}} \times \Delta t}{m} $$


Substituting the known values:


$$ v_{\text{critical}} = \frac{1 \times 10^9 \times 0.1}{100,000} $$


$$ v_{\text{critical}} = 989.1 \, \text{m/s} $$


The meteor will be destroyed if its impact velocity exceeds **989.1 m/s**. Therefore, based on the **UTS of 90 MPa**, the meteor cannot withstand an impact velocity greater than **989.1 m/s** without breaking apart.

Since the meteor will be destroyed if the impact velocity exceeds **989.1 m/s**, we simulate a **random explosion** when this threshold is exceeded.

```python
import numpy as np
import matplotlib.pyplot as plt

# Initial conditions
dt = 0.1  # s
g = 9.81  # m/s^2
tmax = 100  # s
m = 100000  # kg
r = 1.87  # m
A = np.pi * (r ** 2)
V = 4/3 * (np.pi) * (r ** 3)
p = m / V
rho = 1.225  # kg/m^3
Cd = 0.47
delta_t = 0.1  # Assumed impact duration
critical_velocity = 1000  # m/s
mu_r = 0.9  # coefficient of rolling friction
d_penetration = 30  # Penetration depth in meters
F_penetration = 2e6  # Penetration force in Newtons

def meteor_impact_simulation(theta):
    x_initial = 0

    # Initial velocities
    vinitial = 20000  # m/s
    cos_theta_rad = np.cos(np.radians(theta))
    sin_theta_rad = np.sin(np.radians(theta))

    # Initial velocity components based on entry angle
    vi_y = vinitial * sin_theta_rad
    vi_x = vinitial * cos_theta_rad

    h_initial = 100000  # m
    time = 0
    iterations = 0

    # Vertical motion loop
    while h_initial > 0:
        drag_force = (Cd * rho * A * vi_y**2) / (2 * m)
        v_new = vi_y + (g - drag_force) * dt
        h_new = h_initial - (v_new * dt)
        
        if h_new < 0:
            h_new = 0

        vi_y = v_new
        h_initial = h_new
        time += dt
        iterations += 1

    # Final vertical impact velocity
    vertical_impact_velocity = vi_y

    # Horizontal motion loop
    for i in range(iterations):
        drag_force_x = (Cd * rho * A * vi_x**2) / (2 * m)
        v_new_x = vi_x - (drag_force_x * dt)
        x_new = x_initial + (v_new_x * dt)
        
        vi_x = v_new_x
        x_initial = x_new

    # Final horizontal impact velocity
    horizontal_impact_velocity = vi_x

    # Calculate total impact velocity
    v_impact = np.sqrt(vertical_impact_velocity**2 + horizontal_impact_velocity**2)

    # Check for meteor destruction
    if v_impact > critical_velocity:
        rolling_distance = "Meteor Destroyed"
    else:
        # Rolling distance calculation
        v_final = 0.1 * v_impact  # final velocity after penetration
        E_initial = 0.5 * m * v_impact**2  # initial kinetic energy
        E_lost_penetration = F_penetration * d_penetration
        E_final = max(0.5 * m * v_final**2 - E_lost_penetration, 0)
        
        # Rolling distance calculation using Euler’s method
        E_kinetic = E_final
        d = 0
        dt_roll = 0.01
        F_friction = mu_r * m * g

        while E_kinetic > 0:
            v = np.sqrt(2 * E_kinetic / m)
            delta_d = v * dt_roll
            d += delta_d
            E_kinetic -= F_friction * delta_d

        rolling_distance = d

    # Calculate impact force
    impact_force = (m * v_impact) / delta_t

    return theta, vertical_impact_velocity, horizontal_impact_velocity, v_impact, x_initial, impact_force, rolling_distance

# Run simulation for each angle and print results
angles = range(1, 89)  # From 1° to 89°
results = [meteor_impact_simulation(angle) for angle in angles]

# Print formatted table
print(f"{'Angle (°)':>10} {'Vertical Impact Velocity (m/s)':>30} {'Horizontal Impact Velocity (m/s)':>35} {'Total Impact Velocity (m/s)':>30} {'Distance Traveled (m)':>25} {'Impact Force (N)':>20} {'Rolling Distance (m)':>25}")
print("-" * 175)
for theta, v_new, horizontal_impact_velocity, v_impact, x_new, impact_force, rolling_distance in results:
    rolling_distance_display = f"{rolling_distance:.2f} m" if isinstance(rolling_distance, float) else rolling_distance
    print(f"{theta:>10} {v_new:>30.2f} {horizontal_impact_velocity:>35.2f} {v_impact:>30.2f} {x_new:>25.2f} {impact_force:>20.2f} {rolling_distance_display:>25}")
```
